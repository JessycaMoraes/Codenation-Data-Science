```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
# Some settings for matplotlib.
%matplotlib inline

from IPython.core.pylabtools import figsize

figsize(12, 8)

sns.set()
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
pd.set_option('display.max_columns',None)
df_train.head(5)
```

# Correlations

```python
df_train_corr = df_train.corr()
df_train_corr
```

```python
#Correlations with the IN_TREINEIRO column
corr_treino = df_train_corr.loc['IN_TREINEIRO']
corr_treino
```

```python
aux = corr_treino[(corr_treino >= 0.25) | (corr_treino <= -0.25)]
aux
```

```python
#List with variables correlated with the IN_TREINEIRO column
variaveis_sel = aux.index
variaveis_sel
```

# Train

```python
df_train = df_train[variaveis_sel]
```

```python
df_train.head()
```

```python
y_train = df_train['IN_TREINEIRO']
X_train = df_train.drop(columns = 'IN_TREINEIRO')
```

# Test

```python
X_test = df_test[X_train.columns]
X_test
```

# Normalize

```python
X_train = ss.fit_transform(X_train) 
X_test = ss.transform(X_test)
```

# Random Forest Classifier model

```python
#Instantiating the Random Forest Classifier model
classifier = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
#Training the model
classifier.fit(X_train, y_train)
```

```python
y_pred_test = classifier.predict(X_test)
```

```python
df_answer = pd.DataFrame()
df_answer['NU_INSCRICAO'] = df_test['NU_INSCRICAO']
df_answer['IN_TREINEIRO'] = y_pred_test
df_answer
```

```python
df_answer.to_csv('answer.csv', index=False, header=True)
```
